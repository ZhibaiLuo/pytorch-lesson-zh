```python
import torch
import numpy as np
```

```python
x1 = torch.unsqueeze(torch.Tensor(np.linspace(-1, 1, 100)), dim = 1)
x2 = torch.unsqueeze(torch.Tensor(np.linspace(5, 10, 100)), dim = 1)
x = torch.cat((x1, x2), dim = 1)
y = 123 * x1 + 456 * x2 + torch.randn((100, 1))
```

```python
model = torch.nn.Linear(2, 1)
MSE = torch.nn.MSELoss()
adam = torch.optim.Adam(model.parameters(), lr = 5)
epochs = 1000
```

```python
for epoch in range(epochs):
    
    preds = model(x)
    loss = MSE(preds, y)
    adam.zero_grad()
    loss.backward()
    adam.step()
    print('In epoch {}, MSE loss is: {}'.format(epoch, loss))
```
